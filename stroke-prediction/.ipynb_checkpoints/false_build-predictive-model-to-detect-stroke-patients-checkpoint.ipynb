This analysis is going to use logistic regression to predict stroke. It will include 3 parts:
1. Preprocessing data
2. Build Predictive models
3. Predict based on models
4. Determine the best model
5. Conclusion

```python
## Importing packages


library(tidyverse) 
library(tidyverse)
library(caret)
library(lattice)
library(ggplot2)
library(AUC)


## Reading in files

list.files(path = "../input", recursive = TRUE)
train_data <- read_csv("../input/healthcare-dataset-stroke-data/train_2v.csv")
test_data <- read_csv("../input/healthcare-dataset-stroke-data/test_2v.csv")
```

```python
#1. PREPROCESSING DATA
#CONVERT DEPENDENT VAR INTO CHARACTER VAR
train_data$stroke <- ifelse(train_data$stroke==1, 'Yes', 'No')

#CONVERT IT INTO FACTOR VAR
train_data$stroke <- factor(train_data$stroke)
```

Stroke variable has been converted to dummy variable. Therefore, I want to convert to character variable again

```python
#Read train_data
head(train_data,4)
```

Table above shows the dataset includes numerical vars and factor vars. So We need to convert all factor vars into numerical vars, except dependent var (stroke)

```python
dummyvars <- dummyVars(~., data = train_data[,-grep('stroke', names(train_data))], fullRank = TRUE)
data1 <- data.frame(predict(dummyvars, train_data))
head(data1,4)
```

```python
data1$stroke <- train_data$stroke
head(data1,4)
```

```python
#check correlation between all predictor vars
library(corrplot)
cormatrix <- cor(data1[,-17])
head(cormatrix,4)
```

```python
corrplot.mixed(cormatrix,
               lower = "number", upper = "pie", 
               tl.col = "black",tl.pos = "lt")
```

```python

data2 <- data1[,-grep('id|Residence_type.Urban', names(data1), ignore.case=T)]
data2 <- data.frame(data2)
head(data2,4)
```

```python
badCols <- nearZeroVar(data2[,-grep('stroke', names(data2))])
badNames <- colnames(data2)[badCols]
if (length(badCols)>0) {data2 <- data2[,!(names(data2) %in% badNames)] }
```

Dropped variables that are near Zero Variance 

```python
preprocess <- preProcess(data2[,-13], method = c("nzv","center", "scale", "YeoJohnson"))
data3 <- predict(preprocess, data2[,-13])
data3$stroke <- data2$stroke
```

Before starting building models, we need to transform data to be symmetrical 

```python
numberMissing <- NULL
for (j in names(data3)){
  numberMissing <- rbind(numberMissing, c(variable=j,
                                          nmiss=sum(is.na(data3[,j]))))
}
numberMissing
```

```python
# We can see 'bmi''smoking_statusnever.smoked', and 'smoking_statussmokes' varibales have missing values
# So, in order to have accuracy model, we need to impute missing values
# In this case, I'm applying the k-nearest neighbor method to impute missing values

preprocess2 <- preProcess(data3[,-13], method = c('knnImpute'))
preprocess2

data4 <- predict(preprocess2, data3[-13])
data4$stroke <- data3$stroke
head(data4,4)
```

```python
# SPLIT DATASET INTO TRAINING & TESTTING DATA
#Train data is 80%, test data is 20%
set.seed(123)

trainIndex <- createDataPartition(data4$stroke, p=0.8, list=FALSE)

#creating training & testing dataset
traindata <- data4[trainIndex,]
table(traindata$stroke)

testdata <- data4[-trainIndex,]
table(testdata$stroke)
```

```python
# CREATING THE FITCONTROL FOR TRAINING DATA
#  Train a model to predict which products will be returned.
#import parralel processing

library(snow)
library(doSNOW)
library(foreach)
library(iterators)

fitcontrol <- trainControl(method = 'cv', number = 10,
                           classProbs = TRUE,
                           summaryFunction = twoClassSummary)
```

```python
#BUILDING PREDICTIVE MODEL - STOCHASTIC GRADIENT BOOSTING MODEL (method="gbm")
#creating cluster
cl <- makeCluster(3, type = "SOCK")
registerDoSNOW(cl)
set.seed(123)

#creating model

gbmOut <- train(traindata[,-grep('stroke', names(traindata))],
                traindata[, grep('stroke', names(traindata))],
                method = 'gbm', tuneLength = 5,
                trControl = fitcontrol, metric = 'ROC')

stopCluster(cl)
gbmOut
```

```python
#Predict
predictedClass_gbm <- predict(gbmOut, testdata)
predictedProb_gbm <- predict(gbmOut, testdata, type = 'prob')
predictedProb_gbm
```

```python
#Produce Confusion Matrix for gbm

confusionMatrix(data = predictedClass_gbm, reference = testdata$stroke)
```

```python
#Produce Lift Curve for gbm (method="gbm") 
ggplot(lift(testdata$stroke ~ predictedProb_gbm$No, class = 'No'))+
  scale_x_continuous(breaks = seq(0,100,10))+
  scale_y_continuous(breaks = seq(0,100,10))+
  ggtitle('Stochastic Gradient Boosting Lift Curve')
```

```python
#check accuracy with testdata
gbmPred <- predict(gbmOut, testdata)
postResample(gbmPred, testdata$stroke)
```

```python
rocOut_gbm <- roc(predictedProb_gbm$No, testdata$stroke)

auc(rocOut_gbm)
```

```python
rocOut_gbm_yes <- roc(predictedProb_gbm$Yes, testdata$stroke)

auc(rocOut_gbm_yes)
```

```python
#Visualize Model Tuning
ggplot(gbmOut)
```

Tuning parameter 'shrinkage' was held constant at a value of 0.1 Tuning parameter 'n.minobsinnode' was held constant at a value of 10 ROC was used to select the optimal model using the largest value.
The final values used for the model were n.trees = 50, interaction.depth = 3, shrinkage = 0.1 and n.minobsinnode = 10.

```python
#creating cluster
cl <- makeCluster(3, type = "SOCK")
registerDoSNOW(cl)
set.seed(123)

#creating model using Random Forest method
rfmOut <- train(traindata[,-grep('stroke', names(traindata))],
                traindata[, grep('stroke', names(traindata))],
                method = 'rf', tuneLength = 3,
                trControl = fitcontrol, metric = 'ROC')

stopCluster(cl)
```

```python
rfmOut
```

```python
#Predict
predictedClass_rf <- predict(rfmOut, testdata)
predictedProb_rf <- predict(rfmOut, testdata, type = 'prob')
predictedProb_rf
```

```python
#Produce Lift Curve for random forest 
ggplot(lift(testdata$stroke ~ predictedProb_rf$No, class = 'No'))+
  scale_x_continuous(breaks = seq(0,100,10))+
  scale_y_continuous(breaks = seq(0,100,10))+
  ggtitle('Random Forest Lift Curve')
```

```python
#check accuracy with testdata
rfPred <- predict(rfmOut, testdata)
postResample(rfPred, testdata$stroke)
```

```python
#Test ROC with 'no'
rocOut_rf_no <- roc(predictedProb_rf$No, testdata$stroke)

auc(rocOut_rf_no)

#Test ROC with 'yes'
rocOut_rf_yes <- roc(predictedProb_rf$Yes, testdata$stroke)

auc(rocOut_rf_yes)
```

```python
#Visualize Model Tuning
ggplot(rfmOut)
```
